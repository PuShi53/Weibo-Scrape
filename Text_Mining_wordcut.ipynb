```python
#导入必要的工具包
import jieba
import numpy as np
import pandas as pd
from collections import Counter
```

```python
#导入需要切词的数据
df=pd.read_excel('环境_v2.xlsx',sheet_name='Sheet1')
df.head()
```

```python
df_contents=df[['压力词','皮肤词','内容']]
```

```python
df_contents.head()
```

```python
#切词函数
def wordcut(content):
    cut=jieba.cut(content['内容'])
    words=list(cut)
    
    return words
```

```python
#停用词表读取
def stopwordslist(filepath):  
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]  
    return stopwords
```

```python
#searchword_stress=['宝宝','孩子','孕期','父母','爸妈','催婚']
#searchword_stress=['上班','工作','加班','赚钱','公司']
#searchword_stress=['学习','学校','考试','培训','上课','论文','作业']
searchword_stress=['空气','雾霾','季','污染']
searchword_skin=['皮肤','痘','过敏','干燥','皱纹','闭口','黑眼圈','衰老','毛孔','湿疹',
                 '松弛','粉刺','皮炎','黑头','细纹']

stopwords=stopwordslist('停用词整合.txt')

word_count=[]

#切词操作
for i in searchword_stress:
    for j in searchword_skin:
        
        df_for_cut=df_contents[(df_contents['压力词']==i)&(df_contents['皮肤词']==j)]
        
        word_list=[]
        
        for index,item in df_for_cut.iterrows():
            word_list+=wordcut(item)
        
        for word in word_list:
            
            if word not in stopwords:
                
                word_count.append({'压力词':i,'皮肤词':j,'关键词':word})

df_cut=pd.DataFrame(word_count)
df_cut
```

```python
#导出切词结果，后续的词频统计工作在Excel中完成
df_cut.to_excel('环境_切词.xlsx')
```
